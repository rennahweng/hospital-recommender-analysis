# Further Investigations on Duplicates

## 1. Number of Patients vs. "Hospital Service Area"

```python
area_groups = df.groupby(['Hospital Service Area'] )
print( area_groups.count()['Hospital County'] )
```

## 2. "Total Charges" vs. "Hospital Service Area"

```python
ax = area_groups.sum()['Total Charges'].plot(kind='bar', rot=90)
ax.set_ylabel("Total Charges");
```

## 3. Number of Patients vs. Hospital "Facility Name"

```python
name_group = df.groupby(['Facility Name'])
df_name_group = name_group.count()
df_name_group
```

```python
df_name_group.max()
```

```python
df_name_group.min()
```

Since "Total Costs" exists in both min and max, let's just use the "Total Costs" column to get the number of patients for each hospital/facility name.

```python
df_name_group["Total Costs"].describe()
```

```python
# Find the hospital with most number of patients
df_name_group[ df_name_group["Total Costs"] == df_name_group["Total Costs"].describe().max() ]
```

Conclusion: Mount Sinai Hospital has the most number of patients - 52335 records

```python
# Find the hospital with min number of patients
df_name_group[ df_name_group["Total Costs"] == df_name_group["Total Costs"].describe().min() ]
```

Conclusion: Monroe Community Hospital has the least number of patients - 2 records

## 4. Number of Hospitals in Each ZipCode

```python
zipcode_group = df.groupby(['Zip Code - 3 digits'])
df_zipcode_group = zipcode_group.count()
df_zipcode_group
```

```python
df_zipcode_group.max()
```

```python
df_zipcode_group.min()
```

```python
df_zipcode_group["Total Costs"].describe()
```

```python
# Find the zipcode with most number of patients
df_zipcode_group[ df_zipcode_group["Total Costs"] == df_zipcode_group["Total Costs"].describe().max() ]
```

```python
# Find the zipcode with least number of patients
df_zipcode_group[ df_zipcode_group["Total Costs"] == df_zipcode_group["Total Costs"].describe().min() ]
```
